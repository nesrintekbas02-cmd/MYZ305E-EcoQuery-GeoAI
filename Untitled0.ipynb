```python
import os
from getpass import getpass

# API anahtarını güvenli bir şekilde girmek için bir kutucuk açar
print("Lütfen kopyaladığınız Gemini API anahtarınızı yapıştırın ve Enter'a basın:")
os.environ["GEMINI_API_KEY"] = getpass()
```

```python
from langchain.tools import tool
from langchain_google_genai import ChatGoogleGenerativeAI
from langchain_classic.agents import AgentExecutor, create_tool_calling_agent
from langchain_core.prompts import ChatPromptTemplate
import ee

# 1. Ajanın kullanacağı aracı (Tool) tanımlıyoruz
@tool
def landsat_bitki_ortusu_analizi(enlem: float, boylam: float, yil: int) -> str:
    """Belirtilen koordinat (enlem, boylam) ve yıl için Landsat 8 uydu verilerini kullanarak
    yaz aylarındaki ortalama bitki örtüsü yoğunluğunu (NDVI) hesaplar."""
    try:
        nokta = ee.Geometry.Point([boylam, enlem])

        koleksiyon = (ee.ImageCollection('LANDSAT/LC08/C02/T1_TOA')
                      .filterBounds(nokta)
                      .filterDate(f'{yil}-06-01', f'{yil}-09-01')
                      .filterMetadata('CLOUD_COVER', 'less_than', 10))

        resim = koleksiyon.median()
        ndvi = resim.normalizedDifference(['B5', 'B4'])
        deger = ndvi.reduceRegion(reducer=ee.Reducer.mean(), geometry=nokta, scale=30).get('nd').getInfo()

        if deger is None:
            return "Bu bölge ve tarih için uygun bulutsuz uydu görüntüsü bulunamadı."

        return f"Hesaplanan NDVI değeri: {round(deger, 3)}. (Not: 0.2 altı cansız/toprak, 0.2-0.5 zayıf bitki, 0.5 üstü sağlıklı bitki örtüsünü ifade eder.)"

    except Exception as e:
        return f"Analiz sırasında bir hata oluştu: {str(e)}"

# 2. BURAYI DEĞİŞTİRDİK: En güncel model olan 2.5-flash'ı kullanıyoruz
llm = ChatGoogleGenerativeAI(model="gemini-2.5-flash", temperature=0)

# 3. Ajanın kişiliğini ve görevini tanımlıyoruz
prompt = ChatPromptTemplate.from_messages([
    ("system", "Sen EcoQuery adında uzman bir GeoAI ajanısın. Görevin, sana verilen araçları kullanarak kullanıcının coğrafi sorularını yanıtlamaktır. Sonuçları her zaman Türkçe, profesyonel ve anlaşılır bir dille açıkla."),
    ("human", "{input}"),
    ("placeholder", "{agent_scratchpad}"),
])

# 4. Ajanı inşa ediyoruz
araclar = [landsat_bitki_ortusu_analizi]
ajan = create_tool_calling_agent(llm, araclar, prompt)
eco_query_ajani = AgentExecutor(agent=ajan, tools=araclar, verbose=True)

print("EcoQuery Ajanı (Gemini 2.5 Flash modeliyle) başarıyla oluşturuldu ve göreve hazır!")
```

```python
# Ajanımıza sorumuzu soruyoruz
soru = "Kahramanmaraş merkez (enlem: 37.57, boylam: 36.92) civarında 2023 yılı yaz aylarında bitki örtüsü nasıldı? Analiz eder misin?"

print(f"Kullanıcı Sorusu: {soru}\n")
print("EcoQuery Düşünüyor ve Analiz Ediyor...\n")
print("-" * 50)

# Ajanı çalıştırıyoruz
yanit = eco_query_ajani.invoke({"input": soru})

print("-" * 50)
print(f"\nEcoQuery'nin Yanıtı:\n{yanit['output']}")
```

```python
import ee
ee.Initialize(project='kanaryalar')
```

```python
import geemap

# 1. İnteraktif haritayı oluştur ve Kahramanmaraş'a odakla
harita = geemap.Map(center=[37.57, 36.92], zoom=11)

# 2. Ajanımızın kullandığı aynı Landsat 8 verisini çek
nokta = ee.Geometry.Point([36.92, 37.57])
koleksiyon = (ee.ImageCollection('LANDSAT/LC08/C02/T1_TOA')
              .filterBounds(nokta)
              .filterDate('2023-06-01', '2023-09-01')
              .filterMetadata('CLOUD_COVER', 'less_than', 10))

resim = koleksiyon.median()
ndvi = resim.normalizedDifference(['B5', 'B4'])

# 3. Görselleştirme (Renk) parametrelerini ayarla
# Cansız alanlar beyaz/sarı, bitki örtüsü yeşil tonlarında görünecek
ndvi_renkleri = {
    'min': 0.0,
    'max': 0.6,
    'palette': ['#ece7f2', '#d0d1e6', '#a6bddb', '#74a9cf', '#3690c0', '#0570b0', '#034e7b', '#238b45', '#00441b']
}

# 4. Katmanları haritaya ekle
# Sadece Kahramanmaraş merkez etrafında 20km'lik bir alanı kesiyoruz (clip) ki harita çok ağırlaşmasın
analiz_alani = nokta.buffer(20000)

# Önce gerçek renkli (RGB) uydu görüntüsünü altlık olarak ekliyoruz
harita.addLayer(resim.clip(analiz_alani), {'bands': ['B4', 'B3', 'B2'], 'min': 0, 'max': 0.3}, 'Landsat 8 (Gerçek Renk)')

# Üstüne NDVI analizimizi ekliyoruz
harita.addLayer(ndvi.clip(analiz_alani), ndvi_renkleri, 'NDVI Analizi')

# 5. Haritayı ekranda göster
harita
```
